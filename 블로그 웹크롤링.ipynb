# 네이버 블로그 전체

quote 함수를 이용해 아스키 코드 형식이 아닌 글자를 URL인코딩

```python
from urllib.parse import quote
```

```python
query='제주도'
url='https://search.naver.com/search.naver?where=view&sm=tab_jum&query=%EC%A0%9C%EC%A3%BC%EB%8F%84'+quote(query)
```

```python
req=requests.get('https://search.naver.com/search.naver?where=view&sm=tab_jum&query=%EC%A0%9C%EC%A3%BC%EB%8F%84')
html=req.content.decode('utf-8','replace')
soup=BeautifulSoup(html,'html.parser')
```

```python
titles=soup.find_all('li', attrs={'class':'bx _svp_item'})
titles
```

# 블로그 제목 가져오기

```python
for title in titles :
    title=title.find('a', attrs={'class':'api_txt_lines total_tit'}).get_text()
    print('[제목]',title)
```

# 블로그 링크 가져오기

```python
for link in titles :
    link=link.find('a', attrs={'class':'api_txt_lines total_tit'})['href']
    print('[링크]',link)
```
